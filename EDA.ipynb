## Data Preperation

```python
data  = data[['coaster_name', 
#      'Length', 'Speed', 
       'Location', 'Status', 
       # 'Opening date',
       # 'Type', 
       'Manufacturer', 
       # 'Height restriction', 'Model', 'Height',
       # 'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       # 'Track layout', 'Fastrack available', 'Soft opening date.1',
       # 'Closing date', 'Opened', 'Replaced by', 'Website',
       # 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       # 'Single rider line available', 'Restraint Style',
       # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
       # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
       # 'height_value', 'height_unit', 
       'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
       
```

```python
data["opening_date_clean"] = pd.to_datetime(data["opening_date_clean"])
```

```python
data = data.rename(columns= {"coaster_name" : "Coaster_Name", "year_introduced" : "Year_Introduced", "opening_date_clean" : "Opening_Date" , "Inversions_clean" : "Inversions", "Gforce_clean" : "Gforce", "height_ft": "Height_ft", "speed_mph" : "Speed_mph" })
```

```python
data.isna().sum()
```

```python
data.loc[data.duplicated()]
```

```python
data.loc[data.duplicated(subset="Coaster_Name")]
```

```python
# checking an example duplicate
data.loc[ data["Coaster_Name"] == "Crystal Beach Cyclone"]
```

```python
data = data.loc[~data.duplicated(subset= ["Coaster_Name" ,"Location", "Opening_Date"])].reset_index(drop= True).copy()
```

## Feature Understanding

```python
ax = data["Year_Introduced"].value_counts() \
.head(10) \
.plot(kind="bar", title= "Top 10 Years Coasters Introduced", color="skyblue")
ax.set_xlabel("Year Introduced")
ax.set_ylabel("count")
```

```python
# disribution of speeds (mph)
ax  = data["Speed_mph"].plot(kind="hist", bins = 20 , title= " Coaster Speed (mph)")
ax.set_xlabel("speed(mph)")
```

```python
# disribution of speeds (mph)
ax  = data["Speed_mph"].plot.kde()
ax.set_title("Coaster Speed (mph)")
ax.set_xlabel("speed(mph)")
```

## Feature Relationships

## Ask questions about the data

```python
# what are the locations with the fastest roller coasters?

ax = data[data["Location"] != "Other"].groupby("Location")["Speed_mph"].agg(["mean","count"]).query("count >= 10").sort_values(by ="mean")["mean"].plot(kind="barh" , title="Average Coaster Speed by location")
ax.set_xlabel("Average Coaster Speed")
```
